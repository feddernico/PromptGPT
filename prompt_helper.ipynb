```python
import ipywidgets as widgets

from ipywidgets import HTML
from ipywidgets import Layout
from IPython.display import clear_output
from IPython.display import display


# Function to update textarea value
def update_value(change):
    new_subject = change['new']
    prompt_map = {
        'Interpret Code': 'Act as a Data Scientist. Interpret the following code snippet:\n\nENTER CODE SNIPPET',
        'Generate Data': 'Act as a fake data generator. I need a dataset with [NROWS] records and [NCOLS] columns: [COL1], [COL2], ...',
        'Data Science': 'Discuss the following Data Science topic:\n\nENTER TOPIC',
        'Machine Learning': 'Explain the following Machine Learning concept:\n\nENTER MACHINE LEARNING TOPIC',
        'Other': 'ENTER PROMPT'
    }
    input_widget.value = prompt_map.get(new_subject, 'Enter your text here...')

# Dropdown widget for subjects
subject_widget = widgets.Dropdown(
    options=['Interpret Code', 'Generate Data', 'Data Science', 'Machine Learning', 'Other'],
    description='Subject:',
    disabled=False,
)

# Observe changes in subject
subject_widget.observe(update_value, names='value')

# Textarea widget for user input
input_widget = widgets.Textarea(
    placeholder='Enter your text here...',
    description='Input:',
    disabled=False,
    layout=Layout(width='500px', height='200px')
)


# Initialize the placeholder with the correct value
initial_change = {'new': subject_widget.value}
update_value(initial_change)

# Button to generate prompt
generate_button = widgets.Button(
    description='Generate Prompt',
)

# HTML widget to display the prompt
output_widget = HTML(
    value=""
)

# Function to handle button click
def on_button_clicked(b):
    clear_output(wait=True)
    subject = subject_widget.value
    user_input = input_widget.value
    generated_prompt = user_input.replace('\n', '<br />')
    prompt = f"""
<h1>Generated Prompt:</h1>
{generated_prompt}"""
    output_widget.value = prompt

# Set the button click event
generate_button.on_click(on_button_clicked)

# Display the widgets
display(subject_widget, input_widget, generate_button, output_widget)
```
